```python
%matplotlib inline
 
# Import standard packages
import numpy as np
import matplotlib.pyplot as plt

# Import module for the functions
import functions as f
import functions_for_check as fc
```

```python
# settings for plots
plt.rcParams['font.family'] = 'Times New Roman' # font family
plt.rcParams['mathtext.fontset'] = 'stix' # math font
plt.rcParams["font.size"] = 15 # fontsize
plt.rcParams['xtick.labelsize'] = 15 # 
plt.rcParams['ytick.labelsize'] = 15 # 

plt.rcParams['xtick.major.size'] = 10
plt.rcParams['xtick.major.width'] = 1
plt.rcParams['xtick.minor.size'] = 6
plt.rcParams['xtick.minor.width'] = 1

plt.rcParams['ytick.major.size'] = 10
plt.rcParams['ytick.major.width'] = 1
plt.rcParams['ytick.minor.size'] = 6
plt.rcParams['ytick.minor.width'] = 1

plt.rcParams['xtick.direction'] = 'in' # x axis in
plt.rcParams['ytick.direction'] = 'in' # y axis in 
plt.rcParams['axes.linewidth'] = 1.0 # axis line width
plt.rcParams['axes.grid'] = True # make grid
plt.rcParams['grid.linestyle'] = 'dashed'
```

### determine the number of plot points and lmax

```python
div = 50 # number of points between [0, Pi]
lmax = 10 # max value of l which we take into account
epsilon = 1e-4 # regulator
x = np.arange(epsilon, np.pi, (np.pi-2*epsilon)/div) #[epsilon, Pi-epsilon]
```

## $\Gamma^{t,X,zz}_{LM}$ (spin-2 GWs)

### $\Gamma^{t,I,zz}_{LM}$

```python
gamI00 = f.Gamma_I_zz(0,0,x,lmax)
gamI10 = f.Gamma_I_zz(1,0,x,lmax)
gamI11 = f.Gamma_I_zz(1,1,x,lmax)
```

```python
gamI00_5 = f.Gamma_I_zz(0,0,x,5)
gamI10_5 = f.Gamma_I_zz(1,0,x,5)
gamI11_5 = f.Gamma_I_zz(1,1,x,5)
```

```python
gamI00_exa = fc.Gamma_I_zz_exact(0,0,x)
gamI10_exa = fc.Gamma_I_zz_exact(1,0,x)
gamI11_exa = fc.Gamma_I_zz_exact(1,1,x)
```

```python
plt.plot(x/np.pi*180,gamI00,label="$L=0,M=0$",lw=2.5)
plt.plot(x/np.pi*180,gamI10,label="$L=1,M=0$",lw=2.5)
plt.plot(x/np.pi*180,gamI11,label="$L=1,M=1$",lw=2.5)

plt.plot(x/np.pi*180,gamI00_5,color = 'brown',ls = 'dotted',lw=1.5)
plt.plot(x/np.pi*180,gamI10_5,color = 'brown',ls = 'dotted',lw=1.5)
plt.plot(x/np.pi*180,gamI11_5,color = 'brown',ls = 'dotted',lw=1.5)

plt.plot(x/np.pi*180,gamI00_exa,color = 'black', ls = 'dashed')
plt.plot(x/np.pi*180,gamI10_exa,color = 'black', ls = 'dashed')
plt.plot(x/np.pi*180,gamI11_exa,color = 'black', ls = 'dashed')

plt.xlim((0,180))
plt.xticks(np.arange(0, 181, 30))  # put x ticks at intervals of 30

plt.minorticks_on()
plt.gca().xaxis.set_minor_locator(plt.MultipleLocator(10))

plt.xlabel("$\\theta\ [\mathrm{degrees}]$")
plt.ylabel("$\Gamma^{t,I,zz}_{LM}$")
plt.legend(loc='upper right',fontsize = 13)

#plt.savefig("/Users/keisuke/Documents/PTA/pta_and_astrometry/draft_pta_astrometry/draft/figs/gam_t_I_zz_3.pdf", bbox_inches='tight')

plt.show()
```
